```python
import csv

def parseGCode(file):
    currentZ = " "
    currentLine = " "
    with open(file, "r") as g:
        for line in g:
            line = line.strip()  # Remove whitespace
            if len(line) == 0:  # Do not send blank lines
                continue
            if ";" in line:  # Remove inline comments
                line = line.split(";")[0].strip()  # Remove comments starting with ";"
            if len(line) == 0 or line.startswith(";"):  # Don't send empty lines and comments
                continue

            if line.startswith("G1"):  # Strip out "G1" command
                line = line[2:].strip()

            
            if 'Z' in line:
                index = line.index('Z') + 1
                currentZ = line[index:].split()[0]  # Extract the number before the first space
              
            if 'X' in line:
                index = line.index('X') + 1
                number = line[index:].split()[0]  # Extract the number before the first space
                currentLine = number

                if 'Y' in line:
                    index = line.index('Y') + 1
                    number = line[index:].split()[0]
                    currentLine = currentLine + "," + number + "," + currentZ

            if 'Y' in line:
                index = line.index('Y') + 1
                number = line[index:].split()[0]

                if 'X' not in line:
                    if 'Z' not in line:
                        currentLine = "0" + "," + number + ",0"
                    else:
                        currentLine = "0" + " " + number + "," + currentZ
                        print(currentLine)
                    # Add the number to a CSV file

                with open("output.csv", "a", newline="") as csvfile:
                    writer = csv.writer(csvfile)
                    writer.writerow([currentLine])
                    print(currentLine)
                    currentLine = " "
            


    #return lines  # Return the list of G-code lines

file_path = "20mm_calibration.gcode"
parseGCode(file_path)
```
